```python
import csv

with open("/home/axel/Code/Python/championship/goals/csv/homeScorers.csv", newline="", encoding="utf-8") as f:
    reader = csv.reader(f)
    header = next(reader)
```

```python
idx = header.index("assistProfileUrl")

new_header = (
    header[:idx + 1]
    + ["assistPlayerId", "assistKey", "assistInput"]
    + header[idx + 1:]
)

print(len(new_header))  # should be 34
```

```python
df = pd.read_csv(
    "/home/axel/Code/Python/championship/goals/csv/homeScorers.csv",
    skiprows=1,          # skip original header
    names=new_header,    # force 34 columns
    engine="python"      # required
)
```

```python
df.to_csv(
    "/home/axel/Code/Python/championship/goals/csv/NEWhomeScorers.csv",
    index=False,
    encoding="utf-8"
)
```
